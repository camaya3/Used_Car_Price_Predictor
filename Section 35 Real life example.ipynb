## Multicollinearity

```python
data_cleaned.columns.values
```

```python
# We expect Mileage and Year to be correlated. The newer the car, the less miles
```

```python
# One of the vest ways to check for multicollinearity is through VIF(variance inflation facotr)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
# vif - [1,+inf)
# vif = 1: no mulitcollinearity
# 1 < vif < 5: perfectly okay
# vif > 5 : unacceptable

# Year is above 5 so lets go ahead and remove it. Collinearity is too high
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
data_no_multicollinearity
```

## Creating dummy variables

```python
# Pandas has a dummies function n categories n-1 dummies
```

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True, dtype=int)
```

```python
data_with_dummies.head()
```

### Rearrange columns

```python
data_with_dummies.columns.values
```

```python
cols = [ 'log_price','Mileage', 'EngineV','Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear Regression Model

### Declare the inputs and targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

## Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs) #It is not usually recommended to standardize dummy variables
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled,targets, test_size = .2, random_state = 365)
```

## Create the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train) #Log regression. Log transformation on price (y)
```

```python
# A simpole way to check the final result is to plot the predicted values against the observed values
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-y_hat)
plt.title('Residual Plot', size = 18)
```

### R-squared 

```python
reg.score(x_train,y_train)
```

### Find the weight and bias

```python
reg.intercept_ # Bias
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

### Weights interpretation

```python
#1) A postitive weight shows that as a feature increases in value, so do the log_price and 'Price' respectively
```

```python
#2) A negative weight shows that as a feature increases in value, log_price and 'Price' decrease
```

```python
data_cleaned['Brand'].unique() #Audi is the benchmark
```

```python
# Dummy Variables: 1) a postive weight shows that the respective category(brand)
#is more expensive than the benchmark(audi)

#2) A negative weight shows that that the respective category(brand) is less
# expensive than the benchmark (audi)
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test, alpha = .2)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# Our model is good at predicting higher prices. The dots are closer together
# Our model isn't as good as predicting lower prices. The dots are farther away
```

```python
#df_pf = pd.DataFrame(y_hat_test, columns = ['Prediction']) # This gives you the log predictions
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
y_test = y_test.reset_index(drop=True) #We have to do this because y_test kept the original indeces and python is trying to match them. We need to reset them
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target']-df_pf['Prediction']
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target'] * 100)
df_pf
```
